```python
# Play with astronify data
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import astropy.io.fits as fits

from astronify.series import SoniSeries
from astropy.table import Table
```

```python
# Test the document tutorial
data_table = Table({"time":[0, 1, 2, 3, 4, 5, 9, 10, 11, 12],
                    "flux": [0.3, 0.4, 0.3, 0.5, 0.5, 0.4, 0.3, 0.2, 0.3, 0.1]})
```

```python
data_soni = SoniSeries(data_table)
data_soni.note_spacing = 0.2
data_soni.sonify()
data_soni.play()
```

```python
import lightkurve
```

```python
kep12b_lc = lightkurve.search_lightcurvefile("KIC 11804465", cadence="long", quarter=1).download_all()[0].SAP_FLUX.to_table()    
```

```python
test_lc = lightkurve.search_lightcurvefile("KIC 3733346", cadence="long", quarter=1).download_all()[0].SAP_FLUX.to_table()    
```

```python
test_lc
```

```python
# visualize the lightcurve
plt.scatter(kep12b_lc['time'].value, kep12b_lc['flux'], c='k', marker='.',s=1)
plt.xlabel('Time (days)')
plt.ylabel('Flux')
plt.title('Kepler-12b lightcurve')
plt.show()
```

```python
# visualize the lightcurve
plt.scatter(test_lc['time'].value, test_lc['flux'], c='k', marker='.',s=1)
plt.xlabel('Time (days)')
plt.ylabel('Flux')
plt.title('KIC 3733346 lightcurve')
plt.show()
```

```python
kep12b_obj = SoniSeries(kep12b_lc)    
kep12b_obj.sonify()
kep12b_obj.play()
```

```python
# for sonifying stellar spectra 
# need to mix different tones corresponding to different frequencies with different amplitude
# need to generate individual tones for each frequency
data_table = Table({"time":np.arange(10),
                    "flux": np.zeros(10)})

data_soni = SoniSeries(data_table)
data_soni.note_spacing = 0.01
data_soni.center_pitchfloat = 1540
data_soni.sonify()
data_soni.play()
# data_soni.write("test.wav")
```
